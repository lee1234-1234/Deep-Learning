```python
from tensorflow import keras

reuters = keras.datasets.reuters 
(X_train_full, y_train_full), (X_test_full, y_test_full) = reuters.load_data(
    num_words=10000)
```

```python
X_train_full
```

```python
word_index = reuters.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
decoded_newswire = " ".join([reverse_word_index.get(i-3, "?") for i in
    X_train_full[0]])
```

```python
reverse_word_index[1]
```

```python
decoded_review
```

```python
import numpy as np
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

X_train = vectorize_sequences(X_train_full)
X_test = vectorize_sequences(X_test_full)
```

```python
print(X_train_full.shape)
print(X_train.shape)
```

```python
X_train
```

```python
def to_one_hot(labels, dimension=46):
    results = np.zeros((len(labels), dimension))
    for i, label in enumerate(labels):
        results[i, label] = 1.
    return results
    
y_train = to_one_hot(y_train_full)
y_test = to_one_hot(y_test_full)
```

```python
y_train
```

```python
y_train.shape
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=(10000,)))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(46, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
```

```python
X_val = X_train[:1000]
X_train_part = X_train[1000:]
y_val = y_train[:1000]
y_train_part = y_train[1000:]
```

```python
history = model.fit(X_train_part, y_train_part, epochs=30, validation_data=(X_val, y_val))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.DataFrame(history.history).plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 2)
save_fig("annapp_imdb_train_epoch_30")
plt.show()
```

```python
test_result = model.evaluate(X_test, y_test)
```

```python
from tensorflow import keras
housing=keras.datasets.boston_housing
(X_train_full, y_train_full), (X_test_full, y_test_full) = housing.load_data()
```

```python
X_train_full
```

```python
y_train_full
```

```python
X_train = np.copy(X_train_full)
mean = X_train.mean(axis=0)
X_train -= mean
std = X_train.std(axis=0)
X_train /= std

X_test = np.copy(X_test_full)
X_test -= mean
X_test /= std
```

```python
X_train
```

```python
X_test
```

```python
y_train = y_train_full
```

```python
y_test = y_test_full
```

```python
def build_model():
    model = keras.models.Sequential()
    model.add(keras.layers.Flatten(input_shape=(X_train.shape[1],)))
    model.add(keras.layers.Dense(64, activation="relu"))
    model.add(keras.layers.Dense(64, activation="relu"))
    model.add(keras.layers.Dense(1))

    model.compile(loss="mse",
              optimizer="rmsprop",
              metrics=["mae"])

    return model
```

```python
model = build_model()
model.summary()
```

```python
k = 4
num_val_samples = len(X_train) // k
num_epochs = 100
all_scores = []
for i in range(k):
    print(f"Processing fold #{i}")
    
    X_val = X_train[i * num_val_samples: (i + 1) * num_val]
    y_val = y_train[i * num_val_samples: (i + 1) * num_val]
    X_train_part = np.concatenate(
        [X_train[:i * num_val],
         X_train[(i + 1) * num_val:]],
        axis=0)
    y_train_part = np.concatenate(
        [y_train[:i * num_val],
         y_train[(i + 1) * num_val:]],
        axis=0)
    
    model = build_model()
    model.fit(X_train_part, y_train_part,
              epochs=num_epochs, verbose=0)
    
    val_mse, val_mae = model.evaluate(X_val, y_val, verbose=0)
    all_scores.append(val_mae)
```

```python
k = 4
num_val_samples = len(X_train) // k
num_epochs = 200

all_mae_histories = []

for i in range(k):
    print(f"Processing fold #{i}")
    
    X_val = X_train[i * num_val_samples: (i + 1) * num_val]
    y_val = y_train[i * num_val_samples: (i + 1) * num_val]
    X_train_part = np.concatenate(
        [X_train[:i * num_val],
         X_train[(i + 1) * num_val:]],
        axis=0)
    y_train_part = np.concatenate(
        [y_train[:i * num_val],
         y_train[(i + 1) * num_val:]],
        axis=0)
    
    model = build_model()

    history = model.fit(X_train_part, y_train_part,
                        validation_data=(X_val, y_val),
                        epochs=num_epochs, verbose=0)
    
    mae_history = history.history["val_mae"]
    
    all_mae_histories.append(mae_history)
```

```python
all_mae_histories
```

```python
ave_mae_history = [
    np.mean([x[i] for x in all_mae_histories]) for i in range(num_epochs)]
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, len(ave_mae_history) + 1), ave_mae_history)
plt.xlabel("Epochs")
plt.ylabel("Validation MAE")
save_fig("annapp_imdb_train_epoch_30")
plt.show()
```
